```python
def findprime(n):
    primelist = []
    if n == 0 or n == 1:
        primelist = []
    else:
        for i in range(2,n+1):
            if all(i%j!=0 for j in range(2,i)):
                primelist.append(i)
    return primelist
```

```python
def GCD(a,b):
    great = max(a,b)
    less = min(a,b)
    # print(findprime(great))

    if (less == 1) or ((great%less!= 0) and (less in findprime(great))):
        D=1
        # print("หรม",D)
    else:
        for i in range(2,great):
            # print(i)
            # print(great%i)
            if great%i==0:
                D=i
                # print("หรม",D)
                break
    return D
GCD(1,12)
```

```python
a = int(input("ใส่เลขที่จะเป็น a"))
b = int(input("ใส่เลขที่จะเป็น b"))
lista = []
listd = []
listbd =[]
listGD =[]
for i in range(1,a+1):
    if a%i == 0:
        # print(i)
        lista.append(i)
print("d =",lista)
for i in lista:
    if b % i ==0:
        # print(i)
        listd.append(i)
print("d |",b,"=",listd)
for i in listd:
    nB =b/i
    nB =int(nB)
    listbd.append(nB)
print(b,"/d","=",listbd)

for i in listbd:
    index = listbd.index(i)
    # print(listbd[index])
    GD = GCD(lista[index],listbd[index])
    # print(GD)
    if GD == 1:
        listGD.append(lista[index])

ud = max(listGD)
print("unitary divisor ของ ({1},{2}) คือ {0}".format(ud,a,b))
```
