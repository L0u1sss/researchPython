```python
def findprime(n):
    primelist = []
    if n == 0 or n == 1:
        primelist = []
    else:
        for i in range(2,n+1):
            if all(i%j!=0 for j in range(2,i)):
                primelist.append(i)
    return primelist
```

```python
def GMD(a,b):
    great = max(a,b)
    less = min(a,b)
    # print(findprime(great))

    if (great%less!= 0) and (less in findprime(great)) :
        D=1
        # print("หรน",D)
    else:
        for i in range(2,great):
            # print(i)
            # print(great%i)
            if great%i==0:
                D=i
                # print("หรน",D)
                break
    return D
GMD(7,49)
```
